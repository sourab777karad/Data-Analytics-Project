```python
#import necessary libraries
import pandas as pd
import json
```

```python
with open('t20_json_files\\t20_wc_match_results.json') as f:
  data = json.load(f)
  
#Creating a dataframe from the json file
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

```python
#Rename the columns so that we connect the other tables in the other files while working with power bi
df_match.rename({'scorecard' : 'match_id'}, axis=1, inplace=True)
df_match.head()
```

```python
#creating a match_id's dictornary to be used in the other files
match_ids_dict = {}

for index, row in  df_match.iterrows():
    key1 = row['team1'] + ' vs ' + row['team2']
    key2 = row['team2'] + ' vs ' + row['team1']
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']
    
match_ids_dict
    
```

**_Batting Summary_**

```python
with open('t20_json_files\\t20_wc_batting_summary.json') as f:
  data = json.load(f)
  #geting all the records in the single array
  all_records = []
  for record in data:
      all_records.extend(record['battingSummary'])
df_batting  = pd.DataFrame(all_records)
df_batting.head()
```

```python
#changing the dismissal column to out/not out coloumn
df_batting['out/not_out'] = df_batting.dismissal.apply(lambda x: 'out' if len(x) > 0 else 'not out')
df_batting.head()
```

```python
#dropping the dismissal column
df_batting.drop(columns=["dismissal"], inplace=True)
df_batting.head()
```

```python
#removing special characters from the player column
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_batting.head(11)
```

```python
#linking two table by creating one more coloumn related to match_id
df_batting["mach_id"] = df_batting["match"].map(match_ids_dict)
df_batting.head()
```
